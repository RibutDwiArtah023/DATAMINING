**- Definisi:**

Diskritisasi yaitu proses mengubah data bertipe numerik menjadi data bertipe kategorikal.




**- Metode:**

Diskritisasi memiliki 2 pendekatan yaitu *equal-width intervals* dan *equal-frequency intervals*

  *   Equal-width intervals : dikritisasi yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar. 
  *   Equal-frequency intervals : dikritisasi yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota per kelompok yang kurang lebih sama besar.
      
Diskritisasi dengan basis Entropy

 * Entropy digunakan untuk menentukan yang manakah node yang akan
menjadi pemecah data latih berikutnya

###**Soal 2**
**Lakukan proses deskritisasi dengan basis entropy**

**Entropy**

Entropy adalah nilai informasi yang menyatakan ukuran ketidakpastian (impurity) dari attribut dari suatu kumpulan obyek data dalam satuan bit


*   Rumus Entropy:



$$Entropy (D_{1}) = -\sum_{i=1}^{m} Pi \log_{2} Pi$$

*   Rumus Gain:



$$Gain(E_{new}) = E_{initial} - E_{new}$$
*   Rumus Info:



$$Info_{A}(D) = \frac{|D_{_{1}}|}{|D|} Entropy(D_{1}) + \frac{|D_{2}|}{|D|} Entropy(D_{2})$$

Membuat sampel yang akan dianalisis/dihitung


```python
sample = df[["petal_width","sepal_width","sepal_length","petal_length"]]
sample.describe()
```

Membuat kategori random untuk data sampel

```python
np.random.seed(0)
sample["category"] = np.where(np.random.choice(2, sample.shape[0]) < 1, "A", "B")
sample
```

Membuat fungsi getOverCategory yang digunakan untuk menghitung data keseluruhan yang nantinya digunakan untuk menghitung entropy

```python
def getOverCategory(col):
  group = sample.loc[:, :].groupby("category").count()
  a = group.loc["A", col]
  b = group.loc["B", col]
  return (a, b, a+b)
```

Memasukan fungsi splitter digunakan untuk membuat split antara value yang telah ditentukan lalu mengembalikan data yang telah dipisahkan

```python
def splitter(value:float, col:str)->tuple:
  # get data less and greater from value
  less = sample[sample[col] <= value]
  greater = sample[sample[col] > value]

  # calculate into category for each data
  less_group = less.loc[:, :].groupby("category").count()
  greater_group = greater.loc[:, :].groupby("category").count()

  # get value based on category
  less_category_A = less_group.loc["A", col] 
  less_category_B = less_group.loc["B", col] 
  greater_category_A = greater_group.loc["A", col] 
  greater_category_B = greater_group.loc["B", col] 

  return (
      [less_category_A, less_category_B, less_category_A + less_category_B],
      [greater_category_A, greater_category_B, greater_category_A + greater_category_B]
  )
```

Membuat fungsi entropy untuk melakukan perhitungan entropy

```python
def entropy(d):
  r1 = (d[0] / d[2]) * np.log2(d[0] / d[2]) 
  r2 = (d[1] / d[2]) * np.log2(d[1] / d[2]) 
  return np.sum([r1, r2]) * -1
```

Membuat fungsi info

```python
def info(d):
  r1 = (d[0][2] / sample.shape[0]) * entropy(d[0])
  r2 = (d[1][2] / sample.shape[0]) * entropy(d[1])
  return r1 + r2
```

Membuat fungsi gain untuk menghitung selisih antara entropy awal dengan yang baru

```python
def gain(Einitial, Enew):
  return Einitial - Enew
```

**1. Petal Width**


```python
#Membuat Dinitial
a = getOverCategory(("petal_width"))
entropy_a = entropy(a)

print(a)
print(entropy_a)
```

Melakukan beberapa tes split untuk mencari hasil dan informasi yang terbaik 


Tes Pertama : Split 1: 0.1

```python
split1 = splitter(0.1,"petal_width")
info_split1 = info(split1)
gain(entropy_a, info_split1)
```

Tes Kedua : Split 2: 1.3

```python
split2 = splitter(1.3,"petal_width")
info_split2 = info(split2)
gain(entropy_a, info_split2)
```

Tes Ketiga : Split 3: 2.4

```python
split3 = splitter(2.4,"petal_width")
info_split3 = info(split3)
gain(entropy_a, info_split3)
```

Jadi pada percobaan split atribut petal width di dapat bahwa split ke 2 dengan nilai 1.3 merupakan hasil keuntungan informasi terbaik yakni dengan hasil 0.00021232279799998643
